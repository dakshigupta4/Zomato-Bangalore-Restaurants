# <span style='color:Blue'> Import Libraries.  </span>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("dark_background")
```

# <span style='color:Blue'> Reading CSV.  </span>

```python
df=pd.read_csv("zomato.csv")
df.head()#top 5
```

```python
df.shape
```

```python
df.columns
```

```python
df1=df.drop(['url', 'address','menu_item','phone','dish_liked','reviews_list'],axis=1)
df1.head()
```

```python
df1.shape
```

```python
#checking null values
df.info()
```

# Dropping duplicates

```python
#means that there were duplicates value which are now removed
df1.drop_duplicates(inplace=True)
df1.shape
```

# Cleaning

```python
df1["rate"].unique()
```

# <span style ="color:Red">Removing "new", "nan", "-", "/5"</span>

```python
def handlerate(value):
    if(value=="NEW" or value=="-"):
        return np.nan #replacing new and - with null value
    else:
        value=str(value).split("/")
        value=value[0]
        return float(value)#2.2/5 will be return
df1["rate"]=df1["rate"].apply(handlerate)
df1["rate"].head()
```

```python
df1.rate.isnull().sum()
```

# Find null values in rate column (mean)

```python
df1["rate"].fillna(df1["rate"].mean(),inplace= True)#inplace=true because make changes in original dataframe
df1["rate"].isnull().sum()
```

```python
df1.info()
```

# Drop Null Values

```python
df1.dropna(inplace=True)
df1.head()
```

```python
df1.columns
```

```python
df1.rename(columns={"approx_cost(for two people)":"Cost2people","listed_in(type)":"Type"},inplace= True)
df1.head()
```

```python
df1["location"].unique()
```

```python
df1["listed_in(city)"].unique()
```

# We have location and listed_in(city) both but we have to take one thats why we have to drop one

```python
df2=df1.drop(["listed_in(city)"],axis=1)
```

```python
df2.columns
```

```python
df2.head()
```

```python
df2["Cost2people"].unique()
#there is , included like 1,450 1,700   1000 1200
```

# Removing , from cost2people

```python
def handlecomma(value):
    value=str(value)
    if "," in value:
        value=value.replace(",","")
        return float(value)
    else:
        return float(value)
df2["Cost2people"]=df2["Cost2people"].apply(handlecomma)
df2["Cost2people"].unique()
```

```python
df2.head()
```

# Clean rest_type column

```python
df2["rest_type"].unique()
```

```python
rest_types=df["rest_type"].value_counts()
rest_types
```

```python
#rest_type less than 1000
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

# rest_types less than 1k as other

```python
def handle_rest_types(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df2["rest_type"]=df2["rest_type"].apply(handle_rest_types)
df2["rest_type"].value_counts()
#others 9003
```

```python
df2.head()
```

# Cleaning location column

```python
location=df["location"].value_counts()
location
```

```python
location_lessthan300=location[location<300]
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return "others"
    else:
        return value
    
df2["location"]=df2["location"].apply(handle_location)
df2["location"].value_counts()
```

```python
df2.head()
```

# clean cuisines column

```python
cuisines=df["cuisines"].value_counts()
cuisines
```

```python
cuisines_lessthan100=cuisines[cuisines<100]
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return "others"
    else:
        return value
df2["cuisines"]=df2["cuisines"].apply(handle_cuisines)
df2["cuisines"].value_counts()
```

```python
df2.head()
```

```python
df2["Type"].value_counts()
```
